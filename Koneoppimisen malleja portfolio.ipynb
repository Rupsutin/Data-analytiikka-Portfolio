```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import numpy as np

df=pd.read_excel('http://taanila.fi/Concrete_Data.xlsx')
df.head(20)
```

```python
sns.pairplot(df, kind='reg')
```

```python
correlation_matrix = df.corr().round(2)
correlation_matrix
```

```python
df['Age (day)'] = np.log(df['Age (day)'])
correlation_matrix
```

```python
df.columns
```

```python
X = df[['Cement (component 1)(kg in a m^3 mixture)',
       'Blast Furnace Slag (component 2)(kg in a m^3 mixture)',
       'Fly Ash (component 3)(kg in a m^3 mixture)',
       'Water  (component 4)(kg in a m^3 mixture)',
       'Superplasticizer (component 5)(kg in a m^3 mixture)',
       'Coarse Aggregate  (component 6)(kg in a m^3 mixture)',
       'Fine Aggregate (component 7)(kg in a m^3 mixture)', 'Age (day)']]
y = df['Concrete compressive strength(MPa, megapascals) ']
```

```python
from sklearn.linear_model import LinearRegression
malli = LinearRegression().fit(X,y)

malli.coef_
```

```python
malli.intercept_
```

```python
malli.score(X,y)
```

```python
plt.scatter(malli.predict(X), malli.predict(X)-y)
plt.hlines(y=0,xmin=0,xmax=100)

plt.xlabel('Ennuste')
plt.ylabel('Poikkeama todellisesta')
```

```python
plt.scatter(y, malli.predict(X))

plt.xlabel('Todellinen vahvuus')
plt.ylabel('Ennuste')
```
